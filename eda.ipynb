```python
import numpy as np
import pandas as pd
import json
```

```python
data = pd.read_parquet('data/ipums_full_count_nyc_census_coded_20210801.parquet')
data['SINCEIMMIG'] = data['YEAR'] - data['YRIMMIG']
data['AGEIMMIG'] = data['AGE'] - data['YEAR'] + data['YRIMMIG']
data
```

```python
var = pd.read_excel('data/Graphs, Variables, Features.xlsx')
var
```

```python
oov = []
for v in var.variable:
    if v not in data.columns:
        oov.append(v)
oov
```

```python
oovr = []
for v in data.columns:
    if v not in var.variable.values:
        oovr.append(v)
oovr
```

```python
oovf = []
for fs in var['filter']:
    if pd.isna(fs):
        continue
    fs = [x.strip() for x in fs.split(',')]
    for f in fs:
        if f not in data.columns:
            oovf.append(f)
oovf = list(set(oovf))
oovf
```

```python
variables = {}
for chart in var.columns[1:-1]:
    variables[chart] = []
    
    for i in range(len(var)):
        v, vbool = var.loc[i, 'variable'], var.loc[i, chart]
        if pd.isna(vbool):
            continue
        elif vbool == 'Y':
            variables[chart].append(v)
        else:
            raise KeyError
variables
```

```python
filters = {}
for i in range(len(var)):
    v, fs = var.loc[i, 'variable'], var.loc[i, 'filter']
    if pd.isna(fs):
        fs = []
    else:
        fs = [x.strip() for x in fs.split(',')]
    filters[v] = fs
filters
```
